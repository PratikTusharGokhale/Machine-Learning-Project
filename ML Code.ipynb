```
import nltk
nltk.download('punkt')
nltk.download('wordnet')

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.feature_extraction.text import TfidfVectorizer

#loading dataset
ds = pd.read_json("train.json") #reading dataset
ts = pd.read_json("test.json")  #reading dataset

ds.ingredients = ds.ingredients.astype('str')
ts.ingredients = ts.ingredients.astype('str')
ds.ingredients = ds.ingredients.str.replace("["," ")
ts.ingredients = ts.ingredients.str.replace("["," ")
ds.ingredients = ds.ingredients.str.replace("]"," ")
ts.ingredients = ts.ingredients.str.replace("]"," ")
ds.ingredients = ds.ingredients.str.replace("'"," ")
ts.ingredients = ts.ingredients.str.replace("'"," ")
ds.ingredients = ds.ingredients.str.replace(","," ")
ts.ingredients = ts.ingredients.str.replace(","," ")


vect = TfidfVectorizer()
features = vect.fit_transform(ds.ingredients) #returns weights(features)
testfeatures = vect.transform(ts.ingredients)

lE = LabelEncoder()
labels = lE.fit_transform(ds.cuisine)


X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2)


print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)


#from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10) #KNN 
knn.fit(X_train, y_train)
y_pred = knn.predict(testfeatures)
print("Test set predictions:\n {}".format(y_pred))

predconv = lE.inverse_transform(y_pred)
otp = pd.DataFrame({'id':ts.id,'cuisine':predconv})

output = otp[['id','cuisine']] 
print(output)
output.to_csv("output.csv",index = False)
```
